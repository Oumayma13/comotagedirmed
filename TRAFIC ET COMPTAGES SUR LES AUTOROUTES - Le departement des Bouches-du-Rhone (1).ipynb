## Valeur aberrante et nulle :
**Valeurs extrmes :**

Une valeur aberrante est une valeur qui s'carte fortement des valeurs des autres observations, anormalement faible ou leve. Les valeurs aberrantes peuvent avoir un effet disproportionn sur les rsultats statistiques, tels que la moyenne, ce qui peut conduire  des interprtations trompeuses. Et cest souvent  cause des erreurs de saisie de donnes ou bien un problme li au procd.
la couverture en terme de capteurs est rarement totale sur le rseau et les mesures peuvent comporter des erreurs ou tre indisponibles dans le cas de panne de capteur. Le recueil des donnes brutes fournit par consquent empreinte
derreurs. Cest la raison pour laquelle des mthodes de qualication (=identication) des donnes aberrantes et de reconstitution des donnes manquantes sont dveloppes, ces mthodes donnent comme rsultat les tests suivants:
- Le test de sur-comptage est effectu en comparant chaque dbit mesur  un seuil de dbit :  priori 6000 vh./heure.

- Le test de sur-vitesse, est effectu en comparant chaque vitesse mesure  une vitesse limite :  priori 160 km/h.

- Le test du taux d'occupation (To), est effectue en identifiant les valeurs nulles ou suprieures  la 100%.

**Valeurs nulles:**

Lorsque le dbit ou la vitesse est nul ou encore le To la voie est vide. Donc une heure sans donne, revient  dire que la voie est vide cette heure ci.

**Utilisation de graphiques pour identifier des valeurs aberrantes :**
 
Nous pouvons aussi trouver les derniers rsultats des sur-comptage et sur-vitesse en identifiant les valeurs aberrantes par des graphes:
